# 3 Make pie charts for different weather conditions in a month

```python
activ = ["Haze", "Fog", "Smoke", "Mist", "Shallow Fog"]
colors = ['g', 'b', 'r', 'grey', 'y']

slices=[]
for i in activ:
    slices.append(len(data[(data['month']==1) & (data['_conds']==i)]))

plt.pie(slices, labels=activ, colors=colors, shadow=True, autopct="%1.1f%%", startangle=90)

plt.legend(loc=(1, 1))
plt.title("JANUARY")
plt.show()
```
